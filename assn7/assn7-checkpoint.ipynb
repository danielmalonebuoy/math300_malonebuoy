# 2.2 
### The following python function $mcintegrate$ takes a function, a lower bound, an upper bound, and a sample size as arguments.  Using the Monte Carlo technique, $mcintegrate2$ finds an approximation for the integral of the inputted function given a random value of $x$ on the interval $[a,b]$.  The variable z provides the width of each partition.  The for loop iterates across $f(x)$ given a random value of $x$ in the interval $[a,b]$, and sigma is the sum of each step in the iteration. Multiplying z by sigma then gives the approximate area under the curve of $f(x)$ - the integral of $f(x)$.

```python
import numpy as np

def f(x):
    return ((x**3)*(np.sin(4*x)))
```

```python
import numpy as np

def mcintegrate(f, a, b, n):
    np.random.seed(1)
    z = (b-a)/n
    sigma = 0
    for i in range(n):
        x = np.random.rand()*(b-a)+a
        sigma += f(x)
    return z*sigma
```

```python
mcintegrate(f,1.0,7.0,10000)
```

# 2.3 
### The following code block takes the $mcintegrate$ function and compares the estimated definite integral to the actual definite integral.  Then, the error difference between the actual and the estimated values are ploted as the value of sample sizes increases.

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt


a = 1.0
b = 7.0

x = Symbol('x')
F = ((x**3)*(sp.sin(4*x)))
actual = integrate(F, (x, a, b))
    
n1 = 1000
n2 = 100000
n3 = (n2//n1)

increment = np.linspace(n1,n2,n3)
error_list = np.empty([int(n3),1])

for i in increment:
    estimate = mcintegrate(f, a, b, int(i))
    error = np.absolute(actual - estimate)
    error_list[int(((i-n1)/n1))] = error

plt.plot(increment, error_list)
plt.ylabel('Error')
plt.xlabel('Number of Samples')
plt.axis([n1, n2, 0, 10])
plt.show()
```

# Further experimentation of  the $mcintegrate$ function

### The previous graph does not adequately show the relative error in the approximation.  The relative error would more precisely demonstrate the accuracy of $mcintegrate$ as compared to the actual integral.  The following code is quite similar to the previous code block; however, it plots the relative error.
### As the sample size greatly increases, the relative error in the area approaches $2.5\%$.  

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

a = 1.0
b = 7.0

x = Symbol('x')
F = ((x**3)*(sp.sin(4*x)))
actual = integrate(F, (x, a, b))
    
n1 = 1000
n2 = 100000
n3 = (n2//n1)

increment = np.linspace(n1,n2,n3)
error_list = np.empty([int(n3),1])

for i in increment:
    estimate = mcintegrate(f, a, b, int(i))
    error = np.absolute(estimate-actual)/actual
    error_list[int(((i-n1)/n1))] = error

plt.plot(increment, error_list)
plt.ylabel('Error %')
plt.xlabel('Number of Samples')
plt.axis([n1, n2, 0, .2])
plt.show()
```
